```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
a=1
b=1
c=1
Ca=0.14        # en mol/L
Va=10.0        # en mL
Cb=0.10        # en mol/L
Veq=Ca*Va/Cb   # en mL
pasVb=0.1      # en mL
nA,nB,nC,nS_A,nS_B=[],[],[],[],[]
#v=[i/10 for i in range(250)]
v=np.linspace(0,25,int(25/pasVb)+1)
```

```python
print(a,b,c)

print(Ca)
print(Va)
```

```python
print(v)
```

```python
for Vb in v:
    if Vb<Veq:
        nA.append(Ca*Va-Cb*Vb*a/b)
        nB.append(0)
        nC.append(c/b*Cb*Vb)
    else:
        nA.append(0)
        nB.append(Cb*Vb-Cb*Veq)
        nC.append(c/b*Cb*Veq)
    nS_A.append(Ca*Va)
    nS_B.append(Cb*Vb)
```

```python
fig,((ax0,ax1,ax2),(ax3,ax4,ax5))=plt.subplots(2,3,figsize=(16,12))
ax0.plot(v,nB)
ax1.plot(v,nS_A)
ax2.plot(v,nC)
ax3.plot(v,nS_B)
ax4.plot(v,nA)

#Titles
ax0.set_xlabel('Vb (mL)',fontsize=13)
ax0.set_ylabel('n (mol)',fontsize=13)
ax0.set_title('Figure 1',fontsize=15)

ax1.set_xlabel('Vb (mL)',fontsize=13)
ax1.set_ylabel('n (mol)',fontsize=13)
ax1.set_title('Figure 2',fontsize=15)

ax2.set_xlabel('Vb (mL)',fontsize=13)
ax2.set_ylabel('n (mol)',fontsize=13)
ax2.set_title('Figure 3',fontsize=15)

ax3.set_xlabel('Vb (mL)',fontsize=13)
ax3.set_ylabel('n (mol)',fontsize=13)
ax3.set_title('Figure 4',fontsize=15)

ax4.set_xlabel('Vb (mL)',fontsize=13)
ax4.set_ylabel('n (mol)',fontsize=13)
ax4.set_title('Figure 5',fontsize=15)
```
